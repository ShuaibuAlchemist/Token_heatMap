```python
import joblib

# To load the DataFrame:
save_path = r'C:\Users\Muhammad\token_heatmap\Data_Base\df_prices_loaded_solana_ripple_bitcoin_ethereum_20250806_142454.joblib'
df_prices_loaded = joblib.load(save_path)
df_prices_loaded.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
corr = df_prices_loaded.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Token Price Correlation Heatmap')
plt.show()
```

## Token Volatility

```python
import numpy as np

# Calculate log returns
returns = np.log(df_prices_loaded / df_prices_loaded.shift(1))

# Compute volatility (standard deviation of log returns) for each
volatility = returns.std()
print(returns.head())
print("Volatility for each token:")
print(volatility)
```

```python
# Calculate percentage change (simple returns)
pct_returns = df_prices_loaded.pct_change()

print("Percentage change (simple returns) for each token:")
print(pct_returns.head())
```

### Token Correlation for Percentage Change

```python
corr = pct_returns.corr()

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Token Price Correlation Heatmap')
plt.show()
```
